 Python, pandas, matplotlib, nltk, wordcloud, google-play-scraper, re, Counter

```python
!pip install google-play-scraper pandas matplotlib nltk wordcloud
!pip install --upgrade pillow
import pandas as pd
import matplotlib.pyplot as plt
from google_play_scraper import reviews, Sort
from nltk.sentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import nltk
from collections import Counter
import re


# NLTK için gerekli veri
nltk.download('vader_lexicon')
```

Yorumlar - Csv

```python

package_name = 'com.nosurrenderstudio.efsanebaskan'  

result, _ = reviews(
    package_name,
    lang='tr',
    country='tr',
    sort=Sort.NEWEST,
    count=200
)


df = pd.DataFrame(result)

df["at"] = pd.to_datetime(df["at"])

df.to_csv("efsane_baskan_yorumlar.csv", index=False, encoding="utf-8-sig")
df_sorted_desc = df.sort_values("at", ascending=False) # tarihe göre yeniden eskiye
```

Duygu Analizi ve Temizleme

```python
sia = SentimentIntensityAnalyzer()

def get_sentiment(text):
    score = sia.polarity_scores(str(text))['compound']
    if score > 0.05:
        return "Pozitif"
    elif score < -0.05:
        return "Negatif"
    else:
        return "Nötr"

df['sentiment'] = df['content'].apply(get_sentiment)

# Emoji ve özel karakterleri temizleme
df['clean_content'] = df['content'].astype(str).apply(lambda x: re.sub(r'[^\w\sığüşöçİĞÜŞÖÇ]', '', x.lower()))
```

Yıldız - Duygu Grafiği

```python
pivot = pd.crosstab(df["score"], df["sentiment"])
pivot.plot(kind="bar", figsize=(8,6))
plt.title("Yıldız Puanı ve Duygu Analizi İlişkisi")
plt.xlabel("Yıldız Puanı")
plt.ylabel("Yorum Sayısı")
plt.show()
```

Wordcloud - Tüm Yorumlar

```python
# Tüm yorumlar
text_all = " ".join(df.content.astype(str))
wordcloud_all = WordCloud(width=800, height=400, background_color="white", colormap="viridis").generate(text_all)
plt.figure(figsize=(10,6))
plt.imshow(wordcloud_all, interpolation="bilinear")
plt.axis("off")
plt.title("Tüm Yorumlar - Kelime Bulutu")
plt.show()


```

Kelime Frekansı

```python
def get_top_words(text_series, top_n=20, stopwords=None):
   
    all_text = " ".join(text_series.astype(str))
    
    all_text = re.sub(r'[^a-zA-ZığüşöçİĞÜŞÖÇ ]', '', all_text.lower())
   
    words = all_text.split()
    
    if stopwords:
        words = [w for w in words if w not in stopwords]
    
    word_counts = Counter(words)
    
    top_words = pd.DataFrame(word_counts.most_common(top_n), columns=['Kelime','Frekans'])
    return top_words
    
   
stopwords = ['ve', 'bu', 'bir', 'ile', 'da', 'de', 'için', 'oyun', 'gibi', 'çok', 'ama', 'oyuncu'] #gereksiz kelimeler
```

Tüm Yorumlar

```python
top_words_all = get_top_words(df['content'], top_n=20, stopwords=stopwords)
print(top_words_all)

top_words_all.plot(kind='bar', x='Kelime', y='Frekans', figsize=(10,6), color='skyblue')
plt.title("Tüm Yorumlar - En Çok Kullanılan 20 Kelime")
plt.xlabel("Kelime")
plt.ylabel("Frekans")
plt.xticks(rotation=45)
plt.show()
```

Pozitif - Nötr Yorumlar 

```python

top_words_pos = get_top_words(df[df['sentiment']=='Pozitif']['content'], top_n=20, stopwords=stopwords)
top_words_pos.plot(kind='bar', x='Kelime', y='Frekans', figsize=(10,6), color='green')
plt.title("Pozitif Yorumlar - En Çok Kullanılan 20 Kelime")
plt.xlabel("Kelime")
plt.ylabel("Frekans")
plt.xticks(rotation=45)
plt.show()
top_words_neutral = get_top_words(
    df[df['sentiment'] == 'Nötr']['content'],
    top_n=20,
    stopwords=stopwords
)

top_words_neutral.plot(
    kind='bar',
    x='Kelime',
    y='Frekans',
    figsize=(10,6),
    color='gray'
)

plt.title("Nötr Yorumlar - En Çok Kullanılan 20 Kelime")
plt.xlabel("Kelime")
plt.ylabel("Frekans")
plt.xticks(rotation=45)
plt.show()
```

Zaman Serisi Analizi – Yorum Trendleri

```python
df_grouped = df.groupby(df["at"].dt.to_period("M")).size()
df_grouped.plot(kind="line", figsize=(10,5), marker="o",linewidth=5)
plt.title("Yorum Sayısı Zaman Serisi")
plt.xlabel("Tarih (Ay)")
plt.ylabel("Yorum Sayısı")
plt.xticks(rotation=45)
plt.show()
```

Güncelleme Sonrası Pozitif Oran Artışı Analizi

```python

update_dates = ["2025-07-01", "2025-09-01"] #guncelleme tarihini seçtik


def sentiment_change_all_updates(df, update_dates):
    results = []
    df['date'] = pd.to_datetime(df['at'])

    # Güncelleme analizi
    for update in update_dates:
        update_date = pd.to_datetime(update)

        before = df[df['date'] < update_date]
        after = df[df['date'] >= update_date]

        before_counts = before['sentiment'].value_counts(normalize=True) * 100
        after_counts = after['sentiment'].value_counts(normalize=True) * 100

        pos_before = before_counts.get('Pozitif', 0)
        pos_after = after_counts.get('Pozitif', 0)
        change = pos_after - pos_before

        results.append({
            "Güncelleme": update,
            "Öncesi_Pozitif(%)": round(pos_before, 2),
            "Sonrası_Pozitif(%)": round(pos_after, 2),
            "Değişim(%)": round(change, 2)
        })

    results_df = pd.DataFrame(results)

    
    grouped = df.groupby(pd.Grouper(key="date", freq="W"))['sentiment'].value_counts(normalize=True).unstack().fillna(0) * 100
    return results_df, grouped  #zaman serisi için


results_df, grouped = sentiment_change_all_updates(df, update_dates)


print("📊 Güncelleme Öncesi ve Sonrası Pozitif Oranları:")
display(results_df)

ax = grouped['Pozitif'].plot(figsize=(10,5), marker="o", linewidth=2, color="green")


for update_date in update_dates:
    ax.axvline(pd.to_datetime(update_date), color="red", linestyle="--") # Güncelleme tarihlerini gösterme

plt.title("Pozitif Yorum Oranı - Tüm Güncellemeler")
plt.ylabel("Pozitif Oran (%)")
plt.xlabel("Tarih")
plt.show()
```
